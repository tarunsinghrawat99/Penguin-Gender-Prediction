# This is a small project for the prediction of penguin's gender

```python
import pandas as pd
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
penguin= sns.load_dataset('penguins')
```

```python
penguin[penguin.isnull().any(axis=1)]
```

```python
penguin.dropna(inplace=True)
penguin
```

```python
island=pd.get_dummies(penguin['island'],drop_first=True)
species=pd.get_dummies(penguin['species'],drop_first=True)
penguin.drop(['island','species'],inplace=True, axis=1)
penguin=pd.concat([penguin,island,species], axis=1)
penguin
```

```python
y=penguin['sex']
x=penguin.drop('sex',axis=1)
trainx,testx,trainy,testy=train_test_split(x,y)
```

```python
training=RandomForestClassifier(n_estimators=10)
training.fit(trainx,trainy)
```

```python
testing=training.predict(testx)
accuracy_score(testing,testy)
```

```python
result=pd.DataFrame({'realvalue':testy, 'pridicted value': testing})
result.head(35)
```
